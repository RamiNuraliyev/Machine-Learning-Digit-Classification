### 2. SVM Definition

 ####  Parameters:

KERNEL=(Linear,Poly,RBF).

Degree =(1,2,..).

Penalty(C=1000,100,...).



```python
#SVM
def svm_def(k,c,deg=1):
    return SVC(C=c,kernel=k,degree=deg)

clf = svm_def('poly',1000,5)
```

### 3. PCA

$x\in R^{784}$ $-> $ $x\in R^{ d}$

parameter d=?

```python
# X^D to X^d
def pca_d(X,d):
    
    pca = PCA(n_components=d)
    X = pca.fit_transform(X)
    return X
```

### 4. Cross Validation

```python
#cross Validation 

def score(clf,X):
    
    scores = cross_val_score(clf,X,y,cv=4)
    return np.mean(scores)
```

## Results SVM

```python

def score_d_deg(X,d,deg):
    
    svm = svm_def('poly',c=1000,deg=deg)    
    X_d = pca_d(X,d)
    
    return score(svm,X_d)

```

```python
# d=128 score_d_deg

score_128_5= score_d_deg(X,128,5)

score_128_3= score_d_deg(X,128,3)
```

## 5. Kmeans

####  Parameters:

 $y_i\in\{0,1,2,...,9\}$
 
 n_clusters=10
 
 init=? $\in\{'k-means++, random,pca_components\}$

```python

kmeans_K =KMeans(init ='k-means++',n_clusters=10,n_init=10)

kmeans_rnd =KMeans(init ='random',n_clusters=10,n_init=10)
```

```python
#train test split
x_trn,x_tst,y_trn,y_tst=train_test_split(X,y,test_size=0.20)
```

```python


def kmeans_accuracy(kmeans,x_trn,x_tst,y_trn,y_tst,d):
    
    
    #x^D to x^d
    pca = PCA(n_components=d)
    
    x_trn_newD= pca.fit_transform(x_trn)
    
    x_tst_newD= pca.transform(x_tst)
    
#learn classifier

    kmeans.fit(x_trn_newD)

#index i = number(0...9), Y_cluster[i] = Cluster num : digit = 0 cluster = 5 => Y_cluster[0]=5 

    Y_cluster=np.zeros(10)
    for i in range(10):
        Y_cluster[i] = np.bincount(kmeans.labels_[y_trn==i]).argmax()
    
    print("Y_cluster = ",Y_cluster,"\n")
         
    miss = np.sum(kmeans.predict(x_tst_newD) != Y_cluster[y_tst.astype(int)])
    
    print('# misclassified test examples: ',miss,"\n #accuracy: ",1-miss/y_tst.size)
    
    return 1-miss/y_tst.size #accuracy
```

```python
#d = 8

kmeans_k_acc8 = kmeans_accuracy(kmeans_K,x_trn,x_tst,y_trn,y_tst,8)
kmeans_rnd_acc8 = kmeans_accuracy(kmeans_rnd,x_trn,x_tst,y_trn,y_tst,8)
```

```python
#d = 16

kmeans_k_acc16 = kmeans_accuracy(kmeans_K,x_trn,x_tst,y_trn,y_tst,16)
kmeans_rnd_acc16 = kmeans_accuracy(kmeans_rnd,x_trn,x_tst,y_trn,y_tst,16)
```

```python

#d = 128

kmeans_k_acc128 = kmeans_accuracy(kmeans_K, x_trn,x_tst,y_trn,y_tst,128)
kmeans_rnd_acc128 = kmeans_accuracy(kmeans_rnd, x_trn, x_tst, y_trn, y_tst,128)
```

```python


pca_k = PCA(n_components=10).fit(x_trn)

kmeans_pca = KMeans(init =pca_k.components_,n_clusters=10,n_init=1) #pick centers once from the digits
kmeans_pca.fit(x_trn)

Y_cluster=np.zeros(10)

for i in range(10):
    Y_cluster[i] = np.bincount(kmeans_pca.labels_[y_trn==i]).argmax()
    
print("Y_cluster = ",Y_cluster,"\n")

miss = np.sum(kmeans_pca.predict(x_tst) != Y_cluster[y_tst.astype(int)])
    
print('# misclassified test examples: ',miss,"\n #accuracy: ",1-miss/y_tst.size)
    
kmeans_accuracy_pca = 1-miss/y_tst.size
```

## Results Kmeans

```python

idx = range(7)

datad = [{'d':128,'init': "k-means++" ,'accuracy': kmeans_k_acc128},
         {'d':128,'init': "random" ,'accuracy': kmeans_rnd_acc128},
         {'d':16,'init': "k-means++" ,'accuracy': kmeans_k_acc16},
         {'d':16,'init': "random" ,'accuracy': kmeans_rnd_acc16},
         {'d':8,'init': "k-means++" ,'accuracy': kmeans_k_acc8},
         {'d':8,'init': "random" ,'accuracy': kmeans_rnd_acc8},
         {'d':"-",'init': "pca components" ,'accuracy': kmeans_accuracy_pca}]
s = pd.DataFrame(datad)
print(s)
```

```python
#Visualize
plt.figure(figsize=(20,16))
X_new = pca_d(X,2)
kmeans = KMeans(init='k-means++', n_clusters=10, n_init=10)
kmeans.fit(X_new)

# Plot the decision boundary. For that, we will assign a color to each

x_min, x_max = X_new[:, 0].min() - 1, X_new[:, 0].max() + 1
y_min, y_max = X_new[:, 1].min() - 1, X_new[:, 1].max() + 1

h=0.5

xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Obtain labels for each point in mesh. Use last trained model.
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure(1)
plt.clf()
plt.imshow(Z, interpolation='nearest',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap=plt.cm.Paired,
           aspect='auto', origin='lower')

plt.plot(X_new[:, 0], X_new[:, 1], 'k.', markersize=2,alpha=0.6)
# Plot the centroids as a white X
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1],
            marker='x', s=169, linewidths=3,
            color='w', zorder=10)
plt.title('K-means clustering on MNIST dataset (PCA-reduced data)\n'
          'Centroids are marked with white cross')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
plt.show()
```
